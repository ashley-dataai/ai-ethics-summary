# Student Performance Analysis

This notebook demonstrates basic data cleaning and visualization using Python and Pandas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Attendance (%)': [90, 80, 75, 95, 60],
    'Grade': [88, 76, 79, 92, 70]
}
df = pd.DataFrame(data)
df
```

```python
# Basic statistics
df.describe()
```

```python
# Visualization
plt.figure(figsize=(8, 5))
plt.bar(df['Student'], df['Grade'], color='skyblue')
plt.title('Student Grades')
plt.xlabel('Student')
plt.ylabel('Grade')
plt.ylim(0, 100)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/student_performance_sample.CSV')

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Calculate average score
df['average_score'] = df[['math_score', 'reading_score', 'writing_score']].mean(axis=1)

# Group by gender and show mean scores
avg_by_gender = df.groupby('gender')[['math_score', 'reading_score', 'writing_score']].mean()

# Plot math score distribution
plt.figure(figsize=(8, 5))
df['math_score'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
